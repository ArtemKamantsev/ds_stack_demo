```python
image = cv2.imread('../images/geometry.png', -1)
show(image)
```

```python
h, w = image.shape[:2]
M = cv2.getRotationMatrix2D((w / 2, h / 2), 45, 1.0)
M
```

```python
M += [[0, 0, 50],
      [0, 0, 200]]
M
```

```python
rotated = cv2.warpAffine(image, M, (w + 100, h+400))
show(rotated)
```
