```python
import cv2

from opencv_.utils import show
```

```python
img = cv2.imread('../images/animal.jpg', 0)
show(img)
```

```python
ret, thresh1 = cv2.threshold(img, 135, 255, cv2.THRESH_BINARY)
show(thresh1)
```

```python
ret, thresh2 = cv2.threshold(img, 135, 255, cv2.THRESH_BINARY_INV)
show(thresh2)
```

```python
ret, thresh3 = cv2.threshold(img, 90, 255, cv2.THRESH_TRUNC)
show(thresh3)
```

```python
ret, thresh4 = cv2.threshold(img, 135, 255, cv2.THRESH_TOZERO)
show(thresh4)
```

```python
ret, thresh5 = cv2.threshold(img, 135, 255, cv2.THRESH_TOZERO_INV)
show(thresh5)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 1)
show(th2)
```

```python
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 2)
show(th3)
```
