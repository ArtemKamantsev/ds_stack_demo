```python
import cv2

from opencv_.utils import show
```

```python
img = cv2.imread('../images/geometry.png', -1)
show(img)
```

```python
blur_averaging = cv2.blur(img, (50, 50), borderType=cv2.BORDER_DEFAULT)
show(blur_averaging)
```

```python
blur_averaging = cv2.blur(img, (50, 50), borderType=cv2.BORDER_ISOLATED)
show(blur_averaging)
# observation: border color has changed
```

```python
blur_median = cv2.medianBlur(img, 11)
show(blur_median)
```

```python
blur_median = cv2.medianBlur(img, 31)
show(blur_median)
```

```python
blur_gaussian = cv2.GaussianBlur(img, (5, 5), 5)
show(blur_gaussian)
```

```python
blur_gaussian = cv2.GaussianBlur(img, (51, 51), 5)
show(blur_gaussian)
```

```python
blur_gaussian = cv2.GaussianBlur(img, (51, 51), 11)
show(blur_gaussian)
```

```python
img = cv2.imread('../images/animal.jpg', 1)
show(img)
```

```python
blur_gaussian = cv2.GaussianBlur(img, (11, 11), 3)
show(blur_gaussian)
```

```python
blur_bilateral = cv2.bilateralFilter(img, 50, 100, 75)
show(blur_bilateral)
# observation: borders preserved, image blured
```
