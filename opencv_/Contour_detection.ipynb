```python
import cv2

from opencv_.utils import show
```

```python
img = cv2.imread('../images/geometry.png', 1)
show(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show(img_gray)
```

```python
img_edges = cv2.Canny(img, 100, 300, True)
show(img_edges)
```

```python
def detect_contours(source_image, output_image, mode, method, line_type=cv2.LINE_4):
    contours, hierarchy = cv2.findContours(source_image, mode, method)
    img_with_contours = cv2.drawContours(output_image.copy(), contours, -1, (0, 0, 255), line_type)

    return img_with_contours
```

```python
show(detect_contours(img_gray, img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE))
```

```python
show(detect_contours(img_edges, img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE))
```

```python
show(detect_contours(img_edges, img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE, cv2.FILLED))
show(detect_contours(img_edges, img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE, cv2.LINE_4))
show(detect_contours(img_edges, img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE, cv2.LINE_8))
show(detect_contours(img_edges, img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE, cv2.LINE_AA))
```

```python
show(detect_contours(img_edges, img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE))
show(detect_contours(img_edges, img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE))
show(detect_contours(img_edges, img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE))
show(detect_contours(img_edges, img, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE))
```

```python
img_animal = cv2.imread('../images/animal.jpg', 1)
img_animal_edges = cv2.Canny(img_animal, 100, 300, True)
show(img_animal_edges)
```

```python
show(detect_contours(img_animal_edges, img_animal, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE, 1))
show(detect_contours(img_animal_edges, img_animal, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE, 1))
show(detect_contours(img_animal_edges, img_animal, cv2.RETR_TREE, cv2.CHAIN_APPROX_TC89_L1, 1))
show(detect_contours(img_animal_edges, img_animal, cv2.RETR_TREE, cv2.CHAIN_APPROX_TC89_KCOS, 1))
```
