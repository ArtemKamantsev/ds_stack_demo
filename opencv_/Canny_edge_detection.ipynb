```python
import time

import cv2
import numpy as np

from opencv_.utils import show
```

```python
img = cv2.imread('../images/geometry.png', 0)
edges = cv2.Canny(img, 200, 300, True)
show(img)
show(edges)
```

```python
img = cv2.imread('../images/geometry.png', -1)
edges = cv2.Canny(img, 100, 300, True)
show(img)
show(edges)
```

```python
# Get coordinates of lines
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, 40, minLineLength=30, maxLineGap=30)
for l in lines.reshape((-1, 4)):
    cv2.line(img, (l[0], l[1]), (l[2], l[3]), (0, 0, 255), 2)
show(img)
```

# Capture frames from camera and perform detection on them

```python
# acquire the camera
cap = cv2.VideoCapture(0)

for _ in range(10):
    ret, frame = cap.read()  # reads frames from a camera
    edges = cv2.Canny(frame, 100, 150, True)

    # Display edges in a frame
    show(edges)
    time.sleep(0.1)

# release the camera
cap.release()
```
