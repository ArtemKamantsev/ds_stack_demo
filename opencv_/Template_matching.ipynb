```python
import cv2

from opencv_.utils import show
```

```python
img = cv2.imread('../images/animal.jpg', 1)
show(img)
```

```python
img_roe = img[12:, 30:232]
show(img_roe)
```

```python
h, w = img_roe.shape[:-1]
thickness = 2
res = cv2.matchTemplate(img, img_roe, cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
top_left = (max_loc[0] - thickness // 2, max_loc[1] - thickness // 2)
bottom_right = (top_left[0] + w + thickness // 2, top_left[1] + h - thickness // 2)

cv2.rectangle(img,top_left, bottom_right, (0, 255, 0), 2)
show(img)
```
