```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 10)
y = x ** 2
```

```python
axs: plt.Axes
fig, axs = plt.subplots()
axs.plot(x, y, label='test_data')
axs.legend()
# text could be set in TeX format (or LaTeX)
axs.annotate('$y=x^2$', xy=(5, 25), xytext=(2, 40),
            arrowprops={'facecolor':'white', 'shrink':0}, fontsize=18)
axs.secondary_yaxis('right', )
plt.show()
```
