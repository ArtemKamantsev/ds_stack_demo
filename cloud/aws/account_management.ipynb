# Account

- AWS account has 2 types of identities:
    1. **Root account** – full access to the AWS account
    2. **User accounts** – access within allowed permission set

- There are 2 types of user account creation:
    1. **AWS Identity and Access Management (IAM)** – simpler, username+password, lower security, not user-friendly workflow and role switch
    2. **AWS IAM Identity Center** - recommended, could be integrated with external identity provider, exposes better options for accounts management

- Several AWS accounts could be managed under the same Organization (and this is one of the recommended ways to manage resources)
- It is possible to set up AWS cross-account access mechanisms

## Identity and Access Management (IAM) account

- Exists inside 1 AWS account (with possibility of access to the cross-account role)
- The primary use for IAM users is to give **workloads that can't use IAM roles** the ability to make programmatic requests to AWS services using the API or CLI.


## IAM Identity Center account

- User can have access to multiple AWS accounts
- **IAM role** - is automatically assumed

## Account identifiers

- AWS account ID - 12-digit number. Often used in Amazon Resource Names (ARNs) to highlight the account they belong to.
- Canonical user ID – An alpha-numeric identifier, such as 79a59df900b949e55d96a1e698fbacedfd6e09d98eacf8f8d5218e7cd47ef2be, that is an obfuscated form of the AWS account ID.

Note: to find these ids navigate to **Account name (top right corner) → Security credentials → Account details section**

# Regions

- Regions created before 2019 are available by default
- Other regions could be enabled (in some circumstances they will be used without enablement, for AWS Bedrock, for example)
- Regions are divided into partitions. Partitions have independent instances of AWS Identity and Access Management (IAM). Partitions:
    - aws - commercial partition (regular regions)
    - aws-us-gov - US government cloud regions
    - aws-cn - China regions
- Note: if you enable a new region, IAM resources are copied to that region.


# Policy

**Policy** - defines access rights on Identity or Resource. Identity polices are managed from Account Management, Resource polices are managed from that specific resource management page.

- AWS managed policies - contain pre-defined properties for most common use cases.
- Customer managed polices - allow you to define fine-grained access rights.
- **Permissions boundaries for IAM entities** – we can attach policies to Identity to allow of forbid an access to specific resources for that Identity. But also we can define additional policies in section "Set permission boundaries" for IAM Role. Actual permission the IAM Role will have is an intersection of policies attached to the Role and the Permissions boundaries. **Use cases**:
    - Ensure that in future IAM Role will not have redundant access rights.
    - Delegate permission assignment to another person and ensure security.


# Role

- **Identity** – IAM User, IAM group, IAM Role
- **Role** – identity with specific permissions. Difference with IAM user:
    - Role is not associated with one person, a role is intended to be assumable by anyone who needs it.
    - Role provides temporary (short-term) credentials to access AWS resources.
- **Service-linked role** – during service creation they may create roles (linked to that service) to access required resources. An IAM administrator can view, but not edit the permissions for service-linked roles. Before you can delete service-linked roles you must first delete their related resources. This protects your resources because you can't inadvertently remove permission to access the resources.
- Usage: You sign in to the User account, and you can assume a role by switching from a user to an IAM role. This temporarily sets aside your original user permissions and instead gives you the permissions assigned to the role.
    - When you switch roles, your session lasts for some defined period of time.

# Monitoring

- **AWS CloudTrail** - captures API calls and related events and writes them to S3
- **Amazon EventBridge** - in near-real time captures events like: application availability issues or resource changes. You can write simple rules to indicate which events are of interest to you and which automated actions to take when an event matches a rule.\
- **IAM Access Analyzer** – checks your accounts or AWS organization for external access, internal access, and unused access (roles, users, resources, policies)
