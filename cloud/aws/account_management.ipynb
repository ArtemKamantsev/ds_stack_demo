# Account

- AWS account has 2 types of identities:
    1. **Root account** – full access to the AWS account
    2. **User accounts** – access within allowed permission set

- There are 2 types of user account creation:
    1. **AWS Identity and Access Management (IAM)** – simpler, username+password, lower security, not user-friendly workflow and role switch
    2. **AWS IAM Identity Center** - recommended, could be integrated with external identity provider, exposes better options for accounts management

- Several AWS accounts could be managed under the same Organization (and this is one of the recommended ways to manage resources)
- It is possible to set up AWS cross-account access mechanisms

# Regions

- Regions created before 2019 are available by default
- Other regions could be enabled (in some circumstances they will be used without enablement, for AWS Bedrock, for example)
- Regions are divided into partitions. Partitions have independent instances of AWS Identity and Access Management (IAM). Partitions:
    - aws - commercial partition (regular regions)
    - aws-us-gov - US government cloud regions
    - aws-cn - China regions
- Note: if you enable a new region, IAM resources are copied to that region.


# Roles & Policies

- **Policy** - defines access rights of Identity or Resource.
- **Identity** – User account or Role
- **Role** – set of permissions to access AWS resources.
- Usage: You sign in to the User account and can switch to the Role. This temporarily sets aside your original user permissions and instead gives you the permissions assigned to the role.
    - When you switch roles, your session lasts for some defined period of time.
